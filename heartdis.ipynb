### EDA

The dataset has the following attributes: 
    
    1) age
    2) sex
    3) chest pain type (4 values)
    4) resting blood pressure
    5) serum cholestoral in mg/dl
    6) fasting blood sugar > 120 mg/dl
    7) resting electrocardiographic results (values 0,1,2)
    8) maximum heart rate achieved
    9) exercise induced angina
    10) oldpeak = ST depression induced by exercise relative to rest
    11) the slope of the peak exercise ST segment
    12) number of major vessels (0-3) colored by flourosopy
    13) thal: 0 = normal; 1 = fixed defect; 2 = reversable defect

### Splitting the Features and Target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']    #seperating target and 
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data to training data and test data
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size=0.2, stratify=Y, random_state=2)
#X_train contains all the features of the training data and X_test contains all the feautures of the testing data
#test_size is how much of the data we want as testing data
#stratify essentially divides Y in an even manner so that we dont get only 0's or 1's
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Model Training


#### Logistic Regression

```python
model = LogisticRegression()
#training logistic regression model with training data
```

```python
model.fit(X_train,Y_train)    #tries to find relation between Xtrain and Ytrain
```

### Model Evaluation


```python
#we will be using accuracy scores
X_train_prediction =  model.predict(X_train)
X_train_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data:",X_train_accuracy)
```

```python
#accuracy on test data
X_test_prediction =  model.predict(X_test)
X_test_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on training data:",X_test_accuracy)
```

### Building a Predictive system


```python
input_data = (52,1,0,125,212,0,1,168,0,1,2,2,3)
#changing data from tuple to numpy array
input_data_as_array = np.asarray(input_data)
#reshaping the array; as in we are only predicting for one instance
input_data_reshape = input_data_as_array.reshape(1,-1)

prediction = model.predict(input_data_reshape)
#print(prediction)
```
